```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
import torch

from pytorch_model import Classifier
```

```python
base_dir = '../'
```

```python
df1 = pd.read_csv(base_dir+'data/train_human_metrics.csv')
df2 = pd.read_csv(base_dir+'data/train_GPT_metrics.csv')

len(df1), len(df2)
```

```python
# want to renormlize burstiness metric
log_burst1 = np.log10(df1['met1'].values)

new_log_burst1 = (log_burst1-2.3)/2.3
plt.hist(log_burst1, bins=25, alpha=0.75, label='Human')
plt.hist(new_log_burst1, bins=25, alpha=0.75, label='Human renorm')



log_burst2 = np.log10(df2['met1'].values)

new_log_burst2 = (log_burst2-2.3)/2.3
plt.hist(log_burst2, bins=25, alpha=0.75, label='GPT')
plt.hist(new_log_burst2, bins=25, alpha=0.75, label = 'GPT renorm')

plt.xlabel('log10 burstiness')
plt.ylabel('N essays')
plt.legend()
#print(min(log_burst), max(log_burst), max(log_burst)-min(log_burst))
#print(min(new_log_burst), max(new_log_burst), max(new_log_burst)-min(new_log_burst))
```

```python
plt.hist(df1['met5'], bins=15, alpha=0.75, label='Human')
plt.hist(df2['met5'], bins=15, alpha=0.75, label='GPT')
plt.xlabel('Sentiment Combined')
plt.ylabel('N essays')
plt.legend()
```

```python
metrics_cols = ['met%i'%(i+1) for i in range(5)]
arr_human =  df1[metrics_cols].values
arr_gpt =  df2[metrics_cols].values

# concat
arr = np.concatenate((arr_human, arr_gpt))
# normalize the burstiness metric
arr[:,0] = (np.log10(arr[:,0]) - 2.3) / 2.3
```

---
### Load pytorch tensors

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
%%time
human_tensors = torch.load(base_dir+'data/train_human_tensor.pt')
```

```python
# try to put it on gpu
human_tensors = [t.to(device) for t in human_tensors]
```

```python
%%time
gpt_tensors = torch.load(base_dir+'data/train_GPT_tensor.pt')
```

```python
# try to put it on gpu
gpt_tensors = [t.to(device) for t in gpt_tensors]
```

```python
# get lengths
N_human = len(human_tensors)
N_gpt = len(gpt_tensors)
```

```python
# put it in one list
tensors = human_tensors+gpt_tensors
```

```python
word_counts1 = np.array([t.shape[0] for t in human_tensors])
word_counts2 = np.array([t.shape[0] for t in gpt_tensors])
plt.hist(word_counts1, label='Human')
plt.hist(word_counts2, label='GPT')
plt.xlabel('word count')
plt.ylabel('N essays')
max(word_counts1), max(word_counts2)
```

```python
# first col is GPT, second is human
y1 = torch.cat((torch.zeros((1,N_human) ), torch.ones((1,N_human) ))).T
y2 = torch.cat((torch.ones((1,N_gpt) ), torch.zeros((1,N_gpt) ))).T

y = torch.cat((y1,y2), axis=0)
y = y.to(device)
```

```python
from sklearn.model_selection import train_test_split
inds = np.arange(N_human+N_gpt)
train_inds, test_inds = train_test_split(inds, test_size=0.2, random_state=8675309, shuffle=True)
train_inds, test_inds
```

```python
from torch.nn.utils.rnn import pad_sequence
from torch import nn
```

```python
# define model
model = Classifier(use_LSTM=True)
model.to(device)
# Optimizers specified in the torch.optim package
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
loss = nn.CrossEntropyLoss()
```

```python
%%time
batch_size = 50
batch_num = 0

batch_loss=[]
#while batch_num*batch_size < len(train_inds):
while batch_num*batch_size < 2000:
    
    # Zero your gradients for every batch!
    optimizer.zero_grad()
        
    # define batch indicies
    i0 = batch_num*batch_size
    if i0+batch_size < len(train_inds):
        i1 = i0+batch_size
    else:
        i1 =len(train_inds)
    batch_inds = train_inds[i0:i1]
    y_true = y[batch_inds]
    
    # gotta do some padding to put everything in batches    
    x_text = pad_sequence([tensors[ind] for ind in batch_inds])
    x_text.requires_grad = True
    
    # extract the normalized metrics
    x_met = torch.from_numpy(arr[batch_inds]).float().to(device)
    x_met.requires_grad = True
    
    y_probs = model(x_text,x_met)

    L = loss(y_probs, y_true)
    batch_loss.append(L.item())
    L.backward()
    
    optimizer.step()

    batch_num+=1
```

```python
# plot loss vs batch-number
plt.plot(np.arange(len(batch_loss)), batch_loss)
```
